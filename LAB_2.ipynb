# Laboratorio No. 2 Comunicaciones

- Importación

```python
# importacion de paquetes de produccion
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import wavfile
from playsound import playsound
from scipy.fft import fft, fftfreq
import os
```

- Primer punto

```python
file_path = "data/audio.wav"
samplerate, data = wavfile.read(file_path)
channel = 1
signal = data[:, channel]
# playsound("data/audio.wav")
# os.system("afplay" + file_path)
```

* Gráfica

```python
L = len(data)

t = np.linspace(0,18,L)

figure = plt.figure()
figure.set_figwidth(8)
figure.set_figheight(8)

# channel no. 1
plt.plot(t, data[:,0],'b')
# channel no. 2
plt.plot(t, data[:,1], 'r')

plt.title('m(t) vs t')
plt.xlabel('t(s)')
plt.ylabel('m(t)')
plt.grid(color='green', linestyle = '--')
plt.show()
```

```python
dt = 1/samplerate

yf = fft(data)
M_gk = abs(yf)
M_gk = M_gk[0:L//2]
F = (samplerate/L)*np.arange(0, L//2)

figure = plt.figure()
figure.set_figwidth(8)
figure.set_figheight(8)

# channel no. 1
plt.plot(F, M_gk[:,0],'b')
# channel no. 2
plt.plot(F, M_gk[:,1], 'r')

plt.title('A vs Frecuencia(Hz)')
plt.xlabel('Frecuencia(Hz)')
plt.ylabel('A')
plt.grid(color='green', linestyle = '--')
plt.show()
```

- Segundo Punto

```python
fc = 2000 # Hz
sp = np.cos(2*np.pi*fc*t)
s1 = data[:,0]*sp
s2 = data[:,1]*sp

figure = plt.figure()
figure.set_figwidth(8)
figure.set_figheight(8)

# channel no. 1
plt.plot(t, s1,'b')
# channel no. 2
plt.plot(t, s2, 'r')

plt.title('m(t) vs t')
plt.xlabel('t')
plt.ylabel('m(t)')
plt.grid(color='green', linestyle = '--')
plt.show()
```

```python
y_1 = fft(s1)
M_1 = abs(y_1)
M_1 = M_1[0:L//2]
F1 = (samplerate/L)*np.arange(0, L//2)

y_2 = fft(s2)
M_2 = abs(y_2)
M_2 = M_2[0:L//2]
F2 = (samplerate/L)*np.arange(0, L//2)

# channel no. 1
plt.plot(F1, M_1,'b')
# channel no. 2
plt.plot(F2, M_2, 'r')

plt.title('A vs Frecuencia(Hz)')
plt.xlabel('Frecuencia(Hz)')
plt.ylabel('A')
plt.grid(color='green', linestyle = '--')
plt.show()
```
