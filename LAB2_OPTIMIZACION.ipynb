- Primer Punto

```python
def capicua(data) -> None:
    if int == type(data):
        c1 = 0
        c2 = 0
        valor = [int(x) for x in str(data)]
        d = int(len(valor))
        if d > 1:
            rev = valor[::-1]
            if d%2 == 0:
                for i in range(int(d/2)):
                    if valor[i] == rev[i]:
                        c1 = c1 + 1
            else:
                for i in range(int((d-1)/2)):
                    if valor[i] == rev[i]:
                        c2 = c2 + 1
        else:
            return(0)
        if c1 == d/2 or c2 == (d-1)/2: 
            return(1)
        else:
            return(0)
    else: 
        print("Not accepted data")
```

```python
capicua(66)
```

```python
def cantidad(n):
    c = 0
    if int == type(n):
       for i in range(n): 
            if int(capicua(i)) == 1:
                c = c + 1
    else:
        print("Not accepted data")
    return(c)

start = time.time()
p = cantidad(10000000)
end = time.time()

print(end-start)
```
