- Primer Punto

```python
def capicua(data) -> None:
    if int == type(data):
        c1 = 0
        c2 = 0
        valor = [int(x) for x in str(data)]
        d = int(len(valor))
        if d > 1:
            rev = valor[::-1]
            if d%2 == 0:
                for i in range(int(d/2)):
                    if valor[i] == rev[i]:
                        c1 = c1 + 1
            else:
                for i in range(int((d-1)/2)):
                    if valor[i] == rev[i]:
                        c2 = c2 + 1
        else:
            return(0)
        if c1 == d/2 or c2 == (d-1)/2: 
            return(1)
        else:
            return(0)
    else: 
        print("Not accepted data")
```

```python
def cantidad(n):
    c = 0
    if int == type(n):
       for i in range(n): 
            if int(capicua(i)) == 1:
                c = c + 1
    else:
        print("Not accepted data")
    return(c)

start = time.time()
p = cantidad(10000000)
end = time.time()

print(end-start)
```

- Cuarto Punto

#### A + B + C = P(A, B, C)
R:   
> 2A + 3B + 4C = 80 -> maquina X (1)  
2A + 2B + 3C = 60 -> maquina y (2)  
A > 0, B > 0, C > 0  
* Tomemos en cuenta que n > m, por lo tanto tenemos inf soluciones tomare a C como variable parametrica:  
> C = t, para (1) y (2) con A = 0  
3B + 4t = 80 y 2B + 3t = 60 con R: t < 20
* Reduciendo:
> 3B + 4t - 80 = 0 (3) y 2B + 3t -60 = 0 (4)
* Obtenemos (3) - (4): 
> B + t -20 = 0, entonces B = 20 - t
* Reemplazamos B en (1)  
> 2A + 3(20-t) + 4t = 60 -> 2A + 60 - 3t + 4t =80 ->  A = (20 + t)/2
* Obteniendo A y B con C como variable libre

```python
sym.init_printing()

A, B, C = sym.symbols('A,B,C')
f = sym.Eq(2*A+3*B+4*C,80)
g = sym.Eq(2*A+2*B+3*C,60)

sym.solve([f,g],(A,B,C))
```
