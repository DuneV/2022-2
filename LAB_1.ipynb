```python
def distancer2(P1,P2):
    distance = np.sqrt((P2[0]-P1[0])**2 + (P2[1]-P1[1])**2)
    return distance

N = [[-10,5],
     [-3,20],
     [4.3],
     [15,-20],
     [0,-30]]
SP = [3,3]

viable = True
p = 0
for i in range(len(N)-1):
    if distancer2(SP,N[0])<20:
        p = p + 20
    if p < 80.0:
        viable = False
    else: 
        viable = True
print(viable)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(18, 8))
f = 1000 # 1GHz
T = 1/f
w0 = 2*np.pi/f
Nc = 6


t = np.linspace(0,Nc*T,Nc*25)

s1 = 4*np.sin(2*np.pi*f*t)

s2 = 3*signal.sawtooth(2*np.pi*f*t)

s3 = 5*np.sin(2*np.pi*f*t + np.pi/4)


# plt.subplot(3,1,1)
axs[0].plot(t,s1)
axs[0].set_title('Sen(2pi*1000*t) vs t')
axs[0].set_xlabel('t(s)')
axs[0].set_ylabel('F(t)')

axs[0].grid(color='green', linestyle = '--')

axs[0].axhline(y=0, color='k') 

# plt.subplot(3,1,2)
axs[1].plot(t,s2, 'tab:orange')
axs[1].set_title('Sawtooth vs t')
axs[1].set_xlabel('t(s)')
axs[1].set_ylabel('F(t)')

axs[1].grid(color='green', linestyle = '--')

axs[1].axhline(y=0, color='k') 

# plt.subplot(3,1,3)
axs[2].plot(t,s3, 'tab:red')
axs[2].set_title("Sen(2pi*1000*t+ pi/4) vs t")
axs[2].set_xlabel('t(s)')
axs[2].set_ylabel('F(t)')

axs[2].grid(color='green', linestyle = '--')

axs[2].axhline(y=0, color='k') 
plt.show()
```

```python
Nc = 4
f = 1000 # 1GHz
T = 1/f

w0 = 2*np.pi/f
t = np.linspace(0,Nc*T,Nc*25)

s1 = 4*np.sin(2*np.pi*f*t)

s2 = 3*signal.sawtooth(2*np.pi*f*t)

s3 = 5*np.sin(2*np.pi*f*t + np.pi/4)

s4 = s1 + s2 - s3

figure = plt.figure()
figure.set_figwidth(16)
figure.set_figheight(4)

plt.plot(t,s1,'b', linestyle = '--' )
plt.plot(t,s2,'b', linestyle = '--' )
plt.plot(t,s3,'b', linestyle = '--' )
plt.plot(t,s4,'r')
plt.title('Functions vs t')
plt.xlabel('t(s)')
plt.ylabel('F1(t)-F2(t)-F3(t)')
plt.grid(color='green', linestyle = '--')
plt.show
```
